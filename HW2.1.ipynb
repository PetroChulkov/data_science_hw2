```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from itertools import cycle
%matplotlib inline
```

```python
tmp = pd.read_html("https://uk.wikipedia.org/wiki/%D0%9D%D0%B0%D1%81%D0%B5%D0%BB%D0%B5%D0%BD%D0%BD%D1%8F_%D0%A3%D0%BA%D1%80%D0%B0%D1%97%D0%BD%D0%B8", match='Коефіцієнт народжуваності в регіонах України')
```

```python
table = tmp[0]
table
```

```python
#1
table.head()
```

```python
#2
table.shape
```

```python
#3
table = table = table.replace("—", np.nan)
table
```

```python
#4
table.dtypes
```

```python
#5
new_col = pd.to_numeric(table['2014'])
table['2014'] = new_col
new_col2 = pd.to_numeric(table['2019'])
table['2019'] = new_col2
table.dtypes
```

```python
#6
table.isnull().sum()/len(table)
```

```python
#7
table = table.drop([27])
table
```

```python
#8
table_numeric = table.select_dtypes(include='number').fillna(table.select_dtypes(include='number').mean ())
table = pd.concat([table['Регіон'], table_numeric], axis=1)
table
```

```python
#9
average = table['2019'].mean()
higher_values = table[table['2019'] > average]
```

```python
higher_values[['Регіон', '2019']]
```

```python
#10
highest_2014 = table['2014'].max()
```

```python
result = table[table['2014'] == highest_2014][['Регіон']]
```

```python
result
```

```python
#11
plt.bar(table['Регіон'], table['2019'], color=np.random.rand(len(table['Регіон']),4))
plt.xticks(rotation=90)
plt.xlabel("Регіон", fontsize="small", color="midnightblue")
plt.ylabel("Коефіцієнт", fontsize="small", color="midnightblue")
plt.title("Динаміка народжуваності в регіонах України 2019", fontsize=15)
plt.show()
```

```python
labels = table['Регіон']
data = table['2014']

plt.pie(
    data,
    labels=labels,
    shadow=False,
    autopct="%.2f%%",
    pctdistance=1.15,
    labeldistance=1.55,
)

plt.show()
```
