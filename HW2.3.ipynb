```python
#8
min = table['price'].min()
max = table['price'].max()
mean = table['price'].mean()
median = table['price'].median()
min, max, mean, median
# Відповідь: Максимальна ціна? -> 105
# Відповідь: Мінімальна ціна? -> 0
# Відповідь: Середня ціна? -> 13.1
# Відповідь: Медіанна ціна? -> 11.0
```

```python
#9
# Відповідь: Який рейтинг у датасеті найвищий? -> 4.9
highest_rating = table['user_rating'].max()
highest_rating
```

```python
#10
# Відповідь: Скільки книг мають такий рейтинг? -> 52
table['user_rating'].value_counts()[highest_rating]
```

```python
#11
# Відповідь: Яка книга має найбільше відгуків? -> Where the Crawdads Sing
highest_review_rate = table['reviews'].max()
name = table[table['reviews'] == highest_review_rate]['name']
name, highest_review_rate
```

```python
#12
#Відповідь: З тих книг, що потрапили до Топ-50 у 2015 році, яка книга найдорожча (можна використати проміжний датафрейм)? -> Publication Manual of the American Psychologic
table_2015 = table[table['year'] == 2015]
max_price = table_2015['price'].max()
name1 = table_2015[table_2015['price'] == max_price]['name']
max_price, name1
```

```python
#13
# Відповідь: Скільки книг жанру Fiction потрапили до Топ-50 у 2010 році (використовуйте &)? -> 20
table[(table['genre'] == 'Fiction') & (table['year'] == 2010)]['name'].count()
```

```python
#14
#Відповідь: Скільки книг з рейтингом 4.9 потрапило до рейтингу у 2010 та 2011 роках (використовуйте | або функцію isin)? -> 1
table[table['year'].isin([2010, 2011])]['user_rating'].value_counts()[4.9]
```

```python
#15
#І насамкінець, давайте відсортуємо за зростанням ціни всі книги, які потрапили до рейтингу в 2015 році і коштують дешевше за 8 доларів (використовуйте функцію sort_values).
#Відповідь: Яка книга остання у відсортованому списку? -> Old School (Diary of a Wimpy Kid #10)
table[(table['year'].isin([2015])) & (table['price'] < 8)].sort_values(by='price')
```

```python
#16
# Відповідь: Максимальна ціна для жанру Fiction: 82

# Відповідь: Мінімальна ціна для жанру Fiction: 0

# Відповідь: Максимальна ціна для жанру Non Fiction: 105

# Відповідь: Мінімальна ціна для жанру Non Fiction: 0

ordered_table = table.groupby(['genre'])
price_genres = ordered_table['price'].agg(['min', 'max'])
price_genres
```

```python
#17
# Відповідь: Якої розмірності вийшла таблиця? Відповідь: 248

# Відповідь: Який автор має найбільше книг? Відповідь: Jeff Kinney

# Відповідь: Скільки книг цього автора? Відповідь:12

amount_authors = table.groupby(['author'])['name'].agg(['count'])
amount_authors.shape
amount_authors.sort_values(by="count", ascending=False)
```

```python
#18
# Відповідь: У якого автора середній рейтинг мінімальний? Відповідь: Donna Tartt

# Відповідь: Який у цього автора середній рейтинг? Відповідь: 3.9

avg_rating_author = table.groupby(['author'])['user_rating'].agg(["mean"])
avg_rating_author.shape
avg_rating_author.sort_values(by="mean", ascending=True)
```

```python
#19
# Відповідь: Який автор перший у списку? -> Donna Tartt 
splitted_table = pd.concat([amount_authors, avg_rating_author], axis=1)
splitted_table.sort_values(by=['mean', 'count'], ascending=True)
```

```python
#20
data = table.groupby(['year'])['reviews'].agg(['sum'])
plt.plot(data, linestyle='dotted')
plt.title('Річна кількість оглядів', fontsize=15)
plt.show()
```

```python
data1 = table.groupby(['year'])['user_rating'].agg(['mean'])

data1
plt.bar(
    data1.index,
    data1['mean'],
    color=["b", "r", "y", "g"],
)

plt.xlabel("Year", fontsize="small", color="midnightblue")
plt.ylabel("Rating", fontsize="small", color="midnightblue")
plt.title("Average rating by year", fontsize=15)
plt.show()
```

```python
data2 = ordered_table['price'].agg(['max'])
labels = ['Fiction', 'Non Fiction']
plt.pie(
    data2['max'],
    labels=labels,
    shadow=False,
    autopct=lambda x: '{:.0f}'.format(x*data2['max'].sum()/100),
    pctdistance=1.15,
    labeldistance=1.55,
)
plt.title('Max price for genre', y=-0.25)
plt.show()
```
